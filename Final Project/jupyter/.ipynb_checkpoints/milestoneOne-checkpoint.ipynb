```python
# Data manipulation & visualization libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set Kaggle-style aesthetics
sns.set_theme(style="whitegrid", palette="viridis")
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10
```

```python
# Load Pokedex JSON directly from GitHub
url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
pokedex = pd.read_json(url)

# Flatten nested 'pokemon' key if present
pokemon_df = pd.json_normalize(pokedex['pokemon'])

# Preview dataset
pokemon_df.head()
```

# Normalization Process (From JSON to 3NF)

The original Pokédex dataset was provided as a single JSON document containing an array of Pokémon objects. Each Pokémon object included nested arrays (such as `type`, `weaknesses`, and `next_evolution`), as well as compound string attributes (e.g., `"height": "0.71 m"`). This structure is not normalized and violates the principles of relational database design, leading to redundancy and difficulties in querying. The following steps describe how this data was transformed into a fully normalized relational schema in Third Normal Form (3NF).

## Unnormalized Form (UNF)

In the raw JSON, all Pokémon data exists within a single nested object structure. Each Pokémon entry includes repeating groups such as:

- Multiple types in an array (`"type": ["Grass", "Poison"]`)
- Multiple weaknesses in an array (`"weaknesses": ["Fire", "Ice", "Flying", "Psychic"]`)
- Multiple related Pokémon in the evolution chains (`"next_evolution"`, `"prev_evolution"`)

This violates First Normal Form (1NF) since attributes contain non-atomic values and nested structures.

## First Normal Form (1NF)

To achieve 1NF, all repeating groups and arrays were separated into individual tables, and each attribute was made atomic. Examples:

- `type` → moved to a new Type table with a unique `type_id`
- `weaknesses` → moved to a new Weakness table
- `next_evolution` and `prev_evolution` → combined and represented in an Evolution table with `from_pokemon_id` and `to_pokemon_id`
- `"height": "0.71 m"` → split into a numeric attribute `height_m`
- `"weight": "6.9 kg"` → split into a numeric attribute `weight_kg`

Each Pokémon now has a unique primary key (`pokemon_id`), and every column stores only a single value.

## Second Normal Form (2NF)

After achieving 1NF, we ensured that all non-key attributes depend on the entire primary key rather than a part of it. For tables that originally had composite attributes (like Pokémon having multiple types), we introduced junction tables to properly represent many-to-many relationships:

- `PokemonType(pokemon_id, type_id)`
- `PokemonWeakness(pokemon_id, weakness_id)`

This eliminated partial dependencies. For example, type data now depends solely on the `type_id` key, not on any part of the Pokémon entity.

Additionally, repeating information like `"candy"` and `"egg"` were separated into their own reference tables (Candy and Egg), removing redundancy and ensuring that Pokémon sharing the same candy or egg type point to a single record.

## Third Normal Form (3NF)

Finally, transitive dependencies were removed to reach Third Normal Form. Each non-key attribute now depends only on its table’s primary key, and not on other non-key attributes. For instance:

- Candy and Egg contain only attributes directly describing those entities.
- Pokémon attributes such as `height_m`, `weight_kg`, and `spawn_chance` depend only on `pokemon_id`.
- Relationships such as Pokémon types, weaknesses, and evolutions are handled exclusively through foreign keys.

The resulting schema ensures minimal redundancy, maintains referential integrity through foreign keys, and allows flexible joins for analysis and visualization.

## Functional Dependencies

Below are examples of key functional dependencies derived from the normalized schema:

- `pokemon_id` → `num, name, img_url, height_m, weight_kg, spawn_chance, avg_spawns, spawn_time, candy_id, egg_id`
- `candy_id` → `name, candy_count`
- `egg_id` → `distance_km`
- `type_id` → `type_name`
- `weakness_id` → `weakness_name`
- `pokemon_id, type_id` → *(PokemonType relationship)*
- `pokemon_id, weakness_id` → *(PokemonWeakness relationship)*
- `from_pokemon_id, to_pokemon_id` → *(Evolution relationship)*


```python
# Convert numeric columns
pokemon_df['height_m'] = pokemon_df['height'].str.replace(' m', '').astype(float)
pokemon_df['weight_kg'] = pokemon_df['weight'].str.replace(' kg', '').astype(float)

# Fill missing candy counts with zero
pokemon_df['candy_count'] = pokemon_df['candy_count'].fillna(0).astype(int)

# Derive a single primary type for simpler visuals
pokemon_df['primary_type'] = pokemon_df['type'].apply(lambda x: x[0] if isinstance(x, list) else x)
```

```python
type_counts = pokemon_df['primary_type'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=type_counts.index, y=type_counts.values, palette="viridis")
plt.title("Distribution of Pokémon by Primary Type")
plt.xlabel("Pokémon Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=pokemon_df, x='height_m', y='weight_kg',
                hue='primary_type', alpha=0.8, s=100, edgecolor='white')
plt.title("Height vs Weight by Pokémon Type")
plt.xlabel("Height (m)")
plt.ylabel("Weight (kg)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(pokemon_df['candy_count'], bins=15, kde=True, color='mediumslateblue')
plt.title("Distribution of Candy Counts Required for Evolution")
plt.xlabel("Candy Count")
plt.ylabel("Number of Pokémon")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=pokemon_df, x='primary_type', y='weight_kg', palette="magma")
plt.title("Weight Distribution by Pokémon Type")
plt.xlabel("Pokémon Type")
plt.ylabel("Weight (kg)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Expand weaknesses into one row per weakness
weaknesses_exploded = pokemon_df.explode('weaknesses')
weak_counts = weaknesses_exploded['weaknesses'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=weak_counts.index, y=weak_counts.values, palette="coolwarm")
plt.title("Most Common Pokémon Weaknesses")
plt.xlabel("Weakness Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
